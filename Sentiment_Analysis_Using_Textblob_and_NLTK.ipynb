# **SENTIMENT ANALYSIS**

* Here I am using the TextBlob library to classify the sentences either as positive,negative,neutral.
* The function clean_sentence is used to clean the sentence i.e spelling corrections, removing punctuations.
* The re library is used to remove the unwanted sub-strings.
* sub() function is used to replace occurrences of a particular sub-string with another sub-string. This function takes as input the following: The sub-string to replace. The sub-string to replace with. [+-]?([0-9]*[.])?[0-9]+ these substrings are remove and replaced with null string ''.
* The .correct() with correct the spelling mistakes in the setence.
* Now we loop over the sentences in the sent to remove the punctuations and return a clean sentence.
* And finally we would store our cleaned sentence in result.




```python
def sentiment(filename):
  def clean_sentence(sentence):
    sentence = re.sub(r'[+-]?([0-9]*[.])?[0-9]+','',str(sentence))
    # Fix spelling errors in comments!
    sent = TextBlob(sentence)
    sent.correct()
    clean = ""
    for sentence in sent.sentences:   
        words = sentence.words
        # Remove punctuations
        words = [''.join(c for c in s if c not in punctuation) for s in words]
        words = [s for s in words if s]
        clean += " ".join(words)
        clean += ". "
    return clean
  result = [clean_sentence(x) for x in filename]
  sentiment_scores = []
  for sentence in result:
      line = TextBlob(sentence)
      sentiment_scores.append(line.sentiment.polarity)
      if line.sentiment.polarity >0.:
        print(sentence,': Positive\n')
      elif line.sentiment.polarity <=0.1 and line.sentiment.polarity >=-0.1:
        print(sentence,': Neutral\n')
      else:
        print(sentence,': Negative\n')
```

```python
text = ['It is the best model',
        'It is the worst model',
        'I am cooking dinner',
        'The party was okay',
        'It was totally crap',
        'The idea was not good',
        'Can you keep your mouth shut',
        'Can you talk silently?',
        'You are not good at sports']
res = sentiment(text)
```
